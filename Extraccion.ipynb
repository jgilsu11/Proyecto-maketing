```python
df_mini=pd.DataFrame()

for cat in lista_cat_gen:
    url_cat= f"https://minimalismbrand.com{cat}"
    res_cat = requests.get(url_cat)
    sopa_cat = BeautifulSoup(res_cat.content, "html.parser")

    lista_nom_cat = sopa_cat.findAll("a", {"class" :"product-link"})
    lista_product_cat=[product.get("href") for product in lista_nom_cat]

    df_final=pd.DataFrame()
    lista_producto_tot=[]
    lista_prec_tot=[]
    lista_col_tot=[]
    lista_tall_tot=[]
    lista_opi_tot=[]
    lista_rat_tot=[]
    for lin in tqdm(lista_product_cat):
        url=f"https://minimalismbrand.com{lin}"
        res = requests.get(url)
        sopa = BeautifulSoup(res.content, "html.parser")


        #Coger producto
        lista_prod = sopa.findAll("h1", {"class" :"title"})
        lista_producto=[col.getText().lower().strip() for col in lista_prod]
        lista_producto_tot.append(lista_producto)
        

        #Coger precio
        lista_prec = sopa.findAll("span", {"class" :"current-price theme-money"})
        lista_precio=[prec.getText().strip().replace("€", "") for prec in lista_prec]
        lista_prec_tot.append(lista_precio)


        #Coger color
        lista_col = sopa.findAll("div", {"class" :"option-selector__btns"})
        lista_color=[col.getText().replace("\n", " ").strip().split() for col in lista_col]
        lista_col_tot.append(lista_color)

        #Coger tallas
        fieldset_talla = None
        for fieldset in sopa.find_all("fieldset", class_="option-selector-fieldset"):
            legend = fieldset.find("legend", class_="label")
            if legend and "Talla" in legend.get_text():
                fieldset_talla = fieldset
                break  

        lista_talla = []
        if fieldset_talla:
            for label in fieldset_talla.find_all("label", class_="opt-label"):
                talla = label.get_text(strip=True)
                lista_talla.append(talla)
        lista_tall_tot.append(lista_talla)

        #Coger opinion
        lista_rate = sopa.findAll("span", {"class" :"stamped-badge"})
        lista_opi=[col.getText().replace(" Opiniones","")for col in lista_rate]
        df_opi=pd.DataFrame(lista_opi)
        lista_opi_tot.append(lista_opi)

        #Coger rating
        lista_rating=[col["data-rating"] for col in lista_rate]
        lista_rat_tot.append(lista_rating)

    df_prod=pd.DataFrame(lista_producto_tot, columns=["producto"])
    df_prec=pd.DataFrame(lista_prec_tot, columns=["precio"])
    # df_col=pd.DataFrame(lista_col_tot, columns=["color"])
    # df_tall=pd.DataFrame(lista_tall_tot, columns=["tallas"])
    df_opi=pd.DataFrame(lista_opi_tot, columns=["opiniones"])
    df_rat=pd.DataFrame(lista_rat_tot, columns=["rating"])

    df_final=pd.concat([df_final, df_prod, df_prec, df_tall, df_opi, df_rat], axis=1)

    df_mini=pd.concat([df_mini, df_final], ignore_index=True, axis=0)
```
